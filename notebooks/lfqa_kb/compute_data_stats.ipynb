```python
# read in the jsonl.gz files
import json 
import gzip
import glob

def load_json_from_file(gt_file_patterns):
    data = []
    with gzip.open(gt_file_patterns, 'rt', encoding='utf-8') as f:
      lines = f.readlines()
      for line in lines:
        data.append(json.loads(line))
    return data

file_pattern = "/dccstor/srosent2/primeqa-mengxia/kg/train/eli5-train-kilt-oraclekg_all_ans_*jsonl.gz"
files = glob.glob(file_pattern)
```

```python
kg_vocab_stats = {}
kg_sentence_stats = {}
kg_sentence_stats['0'] = 0

for file in files:
    print(file)
    data = load_json_from_file(file)

    for example in data:
        if len(example['kg_sentences']) == 0:
            kg_sentence_stats['0'] += 1
        else:
            if str(example['kg_sentences'][0]['count']) not in kg_sentence_stats:
                kg_sentence_stats[str(example['kg_sentences'][0]['count'])] = 0
            kg_sentence_stats[str(example['kg_sentences'][0]['count'])] += 1
        if str(len(example['kg_vocab'])) not in kg_vocab_stats:
            kg_vocab_stats[str(len(example['kg_vocab']))] = 0
        kg_vocab_stats[str(len(example['kg_vocab']))] += 1
```

```python
def pretty_print_csv(dict):
    for item in dict:
        print(str(item) + "," + str(dict[item]))
```

```python
pretty_print_csv(kg_sentence_stats)
print("vocab")
pretty_print_csv(kg_vocab_stats)
```
